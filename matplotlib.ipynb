# Implicit Approch

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
# Change the default format for the vector formate
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
# Create a data and graph
x = np.linspace(-1,1,50)
y = x**3
```

```python
plt.plot(x,y)
plt.show()
```

# Explicit methon or Apporch 

```python
# let's create the figur
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y)
plt.show()
```

```python
# let's update the graph add some feactuer 
fig = plt.figure() # creat the figure
ax = fig.add_subplot() # plot the line of x and y axis
ax.plot(x,y, label="graph_1") # create a blue line or a inline graph in graph
ax.plot(y,x, label="graph_2") # this one is orange same as blue
ax.set_title("Explicit methon") # this is the title of the graph
ax.set_xlabel("X_coordinate") # this is the title of the x coordinate
ax.set_ylabel("Y_coordinate") # this is the title of the y coordinate
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.legend() # this will we plot the legend of blue and orange
ax.grid() # this provied the grid in graph
plt.show() # this show the graph in the screeen 
```

# Customization of the graph

```python
# let's change the color of the line in graph by the help of color parameter in plot funtion
fig = plt.figure() 
ax = fig.add_subplot()
ax.plot(x,y, color = "#000000", label="graph_1")
ax.plot(y,x, color = "g",label="graph_2") 
ax.set_title("Explicit methon")
ax.set_xlabel("X_coordinate")
ax.set_ylabel("Y_coordinate") 
ax.legend()
ax.grid()
plt.show()
```

```python
#let's strok the line in the graph with help of the  linewidth parameter
""" self, xdata, ydata, linewidth, linestyle, color, gapcolor, marker, markersize, markeredgewidth,
markeredgecolor, markerfacecolor, markerfacecoloralt, fillstyle, antialiased, dash_capstyle, solid_capstyle, dash_joinstyle, solid_joinstyle, pickradius, drawstyle, markevery, **kwargs
"""
fig = plt.figure()
ax = fig.add_subplot()
ax = plt.plot(x,y,color = "#000000", linewidth = 4)
ax = plt.plot(y,x,color = "r", linewidth = 2)
ax = plt.title("linewidth")
plt.show()
```

```python
# let's change the style of the line
fig = plt.figure()
ax = fig.add_subplot()
ax = plt.plot(x,y,color = "#000000", linewidth = 4, linestyle = "--")
ax = plt.plot(y,x,color = "b", linewidth = 2, linestyle = ":")
ax = plt.title("linestyle")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax = plt.plot(x,y,color = "#000000", linewidth = 4,marker = "o",markerfacecolor = "r", markersize = 3)
ax = plt.plot(y,x,color = "b", linewidth = 2, marker = "s",markerfacecolor = "w", markersize = 3)
ax = plt.title("marker porperties", weight = "bold", fontsize = "18")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax = plt.plot(x,y,color = "#000000", linewidth = 4,marker = "o",markerfacecolor = "r", markersize = 3)
ax = plt.plot(y,x,color = "b", linewidth = 2, marker = "s",markerfacecolor = "w", markersize = 3)
ax = plt.title("Grid porperties", weight = "bold", fontsize = "18") # weight and color, alpha that is that opcity|
ax = plt.grid(linewidth = 0.5, color = "r", alpha = 0.6)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_title("legend porperty")
ax.set_xlabel("X-coordinate")
ax.set_ylabel("y-coordinate")
plt.plot(x,y, label = "graph")
# set legend
ax.legend(loc='best')
plt.show()
```

# Scatter plot's

```python
# let's start the to create a x ang y variable then plot in scatter plot
x1 = np.random.rand(200)
y1 = np.random.rand(200)
```

```python
# scatter funtion use to this type of graph
fig = plt.figure()
ax = fig.add_subplot()
# let's change the color (c ="")
# also chage the edgecolor of the circle's by using the edgecolor = "black"
# And also change the opacity of the circle's by using the (alpha parameter)
ax.scatter(x1,y1, c="gray", edgecolor = "k", alpha = 0.6, marker = "*", s=100)
plt.show()
```

```python
x = np.linspace(-2,2,200)
y = x**2 + np.random.rand(200)
z = x**4
# scatter funtion use to this type of graph
fig = plt.figure()
ax = fig.add_subplot()
# let's change the color (c ="")
# also chage the edgecolor of the circle's by using the edgecolor = "black"
# And also change the opacity of the circle's by using the (alpha parameter)
# for size of the circle in graph we use the s=""
img = ax.scatter(x,y, c=z, edgecolor = "k", alpha = 0.6,s=100,cmap = "jet")
fig.colorbar(img)
plt.show()
```

# Bubble plot

```python
x = np.linspace(0,1,100)
y = np.random.rand(100)
z = x**3
fig = plt.figure()
ax = fig.add_subplot()
im = ax.scatter(x,y, s=300*z, c=z, cmap = "jet",edgecolor = "k")
fig.colorbar(im)
```
